```python
# Import the data from SQL.

import pandas as pd
from sqlalchemy import create_engine

engine = create_engine("postgresql://postgres:123456@localhost:5432/crop_data")

df = pd.read_sql("SELECT * FROM crop_data_cleaned", engine)

df.head()
```

```python
# Encode the datas.

from sklearn.preprocessing import LabelEncoder

encoders = {}

for col in ["Area", "Item"]:
    encoder = LabelEncoder()
    df[col] = encoder.fit_transform(df[col])
    encoders[col] = encoder
```

```python
# Independent variables.

x = df.drop(['Production'],axis = 1)

# Dependent variable.

y = df['Production']
```

```python
x.head()
```

```python
y.head()
```

```python
# Spilt the data for train and test.

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import Ridge, Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, r2_score

# Ridge Model
ridge = Ridge(alpha=1.0)
ridge.fit(x_train, y_train)

ridge_pred = ridge.predict(x_test) 

print("Ridge MAE:", mean_absolute_error(y_test, ridge_pred))
print("Ridge R2:", r2_score(y_test, ridge_pred))


# Lasso Model
lasso = Lasso(alpha=0.1)
lasso.fit(x_train, y_train)

lasso_pred = lasso.predict(x_test)

print("Lasso MAE:", mean_absolute_error(y_test, lasso_pred))
print("Lasso R2:", r2_score(y_test, lasso_pred))

#decision tree
dt = DecisionTreeRegressor()
dt.fit(x_train, y_train)

dt_pred = dt.predict(x_test)

print("Decision Tree MAE:", mean_absolute_error(y_test, dt_pred))
print("Decision Tree R2:", r2_score(y_test, dt_pred))


#random forest
random = RandomForestRegressor()
random.fit(x_train, y_train)

random_pred = random.predict(x_test)

print("random forest MAE:", mean_absolute_error(y_test, random_pred))
print("random forest R2:", r2_score(y_test, random_pred))
```

```python
# Store the model and encoder.

import joblib

joblib.dump(encoders, "encoded_data.pkl")
joblib.dump(random, "random_forest.pkl")
```
